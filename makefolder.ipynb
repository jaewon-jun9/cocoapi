```python
from pycocotools.coco import COCO
import numpy as np
import json
from PIL import Image
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
import os
import shutil
pylab.rcParams['figure.figsize'] = (8.0, 10.0)
%matplotlib inline
```

```python
num_images=10000 #10000번 이미지 까지 자동으로 분류 폴더 만들고 복사
for num in range(1,num_images+1):
    json_name = './dataset/train/annos/' + str(num).zfill(6)+'.json'
    image_name = './dataset/train/image/' + str(num).zfill(6)+'.jpg'
    with open(json_name, 'r') as f:
        temp = json.loads(f.read())
    try:
        i_name1=temp['item1']['category_name']
        print(i_name1)
        print(temp['pair_id'])
    except:
        print('no item1')    
        i_name1='no'
    try:
        i_name2=temp['item2']['category_name']
        print(i_name2)
        print(temp['pair_id'])
    except:
        print('no item2')
        i_name2='no'
    try:
        if not(os.path.isdir('./dataset/kml/'+i_name2+'_'+i_name1)):
            os.makedirs(os.path.join('./dataset/kml/'+i_name2+'_'+i_name1))
        else:
            print('이미존재하는디렉터리')        
    except OSError as e:
        if e.errno != errno.EEXIST:
            print("Failed to create directory!!!!!")
            raise
    shutil.copy(image_name,'./dataset/kml/'+i_name2+'_'+i_name1+'/')
    print(image_name,'./dataset/kml/'+i_name2+'_'+i_name1+'/')
```
